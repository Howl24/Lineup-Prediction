Como los encuentros del mundial aún no cuentan con datos, los removemos del conjunto

```python
df = df.loc[df['Torneo'] != 'World Cup - 2018 Russia']
```

Las filas en las que el número de partidos jugados es igual a 0 tampoco cuentan con información por lo que se remueven del conjunto de datos

```python
df.loc[df['Matches played'] == 0].head()
```

```python
df = df.loc[df['Matches played'] != 0]
```

```python
df
```

```python
team_df = pd.DataFrame()
team_group_sum = df.groupby('Equipo').sum()

team_df['team'] = team_group_sum.index
team_df['matches_played'] = team_group_sum['Matches played'].values
team_df['ratio_wins'] = (team_group_sum['Wins'] / team_group_sum['Matches played']).values
team_df['ratio_losses'] = (team_group_sum['Losses'] / team_group_sum['Matches played']).values
team_df['ratio_draws'] = (team_group_sum['Draws'] / team_group_sum['Matches played']).values
team_df['avg_goals_for'] = (team_group_sum['Goals for'] / team_group_sum['Matches played']).values
team_df['avg_goals_against'] = (team_group_sum['Goals against'] / team_group_sum['Matches played']).values


team_df = team_df.set_index('team')
```

Se reemplazan los nombres de los equipos para compararlos

```python
info = info.replace("Russia", "Rusia")
info = info.replace("France", "Francia")
info = info.replace("Peru", "Perú")
info = info.replace("Brazil", "Brasil")
info = info.replace("Poland", "Polonia")
info = info.replace("Egypt", "Egipto")
info = info.replace("Germany", "Alemania")
info = info.replace("Spain", "España")
info = info.replace("Japan", "Japón")
info = info.replace("Saudi Arabia", "Arabia Saudí")
info = info.replace("England", "Inglaterra")
info = info.replace("Mexico", "México")
info = info.replace("Morocco", "Marruecos")
info = info.replace("Tunisia", "Túnez")
info = info.replace("Belgium", "Bélgica")
info = info.replace("Croatia", "Croacia")
info = info.replace("Denmark", "Dinamarca")
info = info.replace("Iceland", "Islandia")
info = info.replace("Korea Republic", "República de Corea")
info = info.replace("Panama", "Panamá")
info = info.replace("Switzerland", "Suiza")
info = info.replace("Sweden", "Suecia")
info = info.replace("Republic of Iran", "Iran")
info = info.replace("Iran", "RI de Irán")
```

```python
list(team_df.index)
```

```python
info.head()
```

```python
matches = []
for idx, row in info.iterrows():
    match_row = {}
    try:
        team1 = row['EquipoA']    
        team1_attrs = team_df.loc[team1]

        team2 = row['EquipoB']
        team2_attrs = team_df.loc[team2]
    except KeyError:
        continue
    
    
    match_row['team1'] = team1
    match_row['team2'] = team2
    match_row['diff_partidos_jugados'] = team1_attrs['matches_played'] - team2_attrs['matches_played']
    match_row['diff_ratio_wins'] = team1_attrs['ratio_wins'] - team2_attrs['ratio_wins']
    match_row['diff_ratio_draws'] = team1_attrs['ratio_draws'] - team2_attrs['ratio_draws']
    match_row['diff_ratio_losses'] = team1_attrs['ratio_losses'] - team2_attrs['ratio_losses']
    match_row['diff_avg_goals_for'] = team1_attrs['avg_goals_for'] - team2_attrs['avg_goals_for']
    match_row['diff_avg_goals_against'] = team1_attrs['avg_goals_against'] - team2_attrs['avg_goals_against']
    
    
    match_row['diferencia_goles'] = (row['GolesA'] - row['GolesB'])
    if match_row['diferencia_goles'] < 0:
        match_row['resultado'] = 1
    if match_row['diferencia_goles'] == 0:
        match_row['resultado'] = 2
    if match_row['diferencia_goles'] > 0:
        match_row['resultado'] = 3
    
    matches.append(match_row)
matches = pd.DataFrame(matches)
```

```python
X = matches.drop(['diferencia_goles', 'diff_partidos_jugados',  'team1', 'team2', 'resultado'], axis=1)
```

```python
y = matches['resultado']
```

```python
from sklearn.manifold import MDS
```

```python
mds = MDS(n_components=2)
```

```python
X_mds = mds.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
```

```python
X_lose = X_mds[y == 1]
X_draw = X_mds[y == 2]
X_win = X_mds[y == 3]
```

```python
plt.scatter(X_lose[:, 0], X_lose[:, 1], color='navy', lw=0, label='Lose')
plt.scatter(X_draw[:, 0], X_draw[:, 1], color='turquoise', lw=0, label='Draw')
plt.scatter(X_win[:, 0], X_win[:, 1], color='darkorange', lw=0, label='Win')
plt.legend()
```
