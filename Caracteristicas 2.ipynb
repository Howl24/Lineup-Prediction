```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("caracteristicasEquipos.csv")
```

```python
df = df[df['Torneo'] != "World Cup - 2018 Russia"]
```

```python
cols = df.columns.tolist()
```

```python
cols.remove("Equipo")
```

```python
cols.remove("Torneo")
```

```python
cols = ['Equipo', 'Torneo'] + cols
```

```python
cols
```

```python
df = df[cols]
```

```python
df
```

```python
group = df.groupby(['Equipo']).mean()
group_gc = group['Avg. goals conceded p/m'].sort_values()
```

```python
y = range(len(group))
```

```python
group_gc
```

```python
group_gc.keys()
```

```python
plt.figure(figsize=(20,10))
plt.barh(y,group_gc)
plt.yticks(y, group_gc.keys())
plt.xlabel("Prom. goles recibidos por partido")
```

```python
group = df.groupby(['Equipo']).mean()
group_gs = group['Avg. goals scored p/m'].sort_values()
```

```python
plt.figure(figsize=(20,10))
plt.barh(y,group_gc)
plt.yticks(y, group_gs.keys())
plt.xlabel("Prom. goles anotados por partido")
```
